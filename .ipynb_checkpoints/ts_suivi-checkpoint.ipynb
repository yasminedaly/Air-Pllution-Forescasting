```r
Time<-c(1:length(ts_daily))
model1<-lm(ts_daily~Time)
summary(model1)
```

```r
plot(ts_daily,lwd=2)
points(time(ts_daily),fitted(model1),col="orange2",lwd=2,type="l")
```

```r
model2<-lm(ts_daily~Time+I(Time^2))
#help(I)
summary(model2)
```

```r
plot(ts_daily,lwd=2)
points(time(ts_daily),fitted(model1),col=2,lwd=2,type="l")
points(time(ts_daily),fitted(model2),col=3,lwd=2,type="l")
```

```r
T=7
MC=matrix(0,length(ts_daily),6)
MS=matrix(0,length(ts_daily),6)
for (i in 1:(T/2)){MC[,i]=cos(2*pi*i*Time/T)}
for (i in 1:(T/2)){MS[,i]=sin(2*pi*i*Time/T)}
```

```r
model3<-lm(ts_daily~Time+MC+MS)
summary(model3)
```

```r
model4<-lm(ts_daily~Time+I(Time^2)+MC+MS)
summary(model4)
```

```r
model5<-lm(log(ts_daily)~Time+I(Time^2)+MC+MS)
summary(model5)
```

```r
# Models of the document 

t<-c(1:length(ts_daily))
s.t<-(t-mean(t))/sd(t) # this variable is introduced in order to guarantee
```

```r
CS<-matrix(0,length(ts_daily),6)
SN<-matrix(0,length(ts_daily),6)
for (i in 1:3) CS[,i]<-cos(2*pi*i*t/12)
for (i in 1:3) SN[,i]<-sin(2*pi*i*t/12)
```

```r
Model1<-lm(ts_daily~s.t+CS+SN)

summary(Model1)
```

```r
Model2<-lm(ts_daily~s.t+I(s.t^2)+CS+SN)

summary(Model2)
```

```r
# comaprison of Model 1 and Model2

plot(ts_daily,lwd=2)
points(time(ts_daily),fitted(Model1),col=2,lwd=2,type="l")
points(time(ts_daily),fitted(Model2),col=3,lwd=2,type="l")
```

```r
sp.res<-Model2$residuals
plot(time(ts_daily),sp.res, xlab="Time", ylab="",
 main="Residual component",
 col="purple4",lwd=2,type="l")
```

```r
library(tseries)
adf.test(sp.res)
```

```r
acf(sp.res)
Box.test(sp.res)
```

```r
acf(sp.res,50)
pacf(sp.res,50)
```

```r
# our best model is  ARIMA (0 , 1 , 3)

Model.res<-arima(sp.res,order = c(1, 0, 1))
adf.test(Model.res$residuals)
checkresiduals(Model.res)
```

```r
d.sp.ts.24<-diff(sp.res,24,1)
acf(d.sp.ts.24,50)
pacf(d.sp.ts.24,50)
```
