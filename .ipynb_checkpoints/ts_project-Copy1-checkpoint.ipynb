```r
install.packages("zoo")
install.packages("xts")
library(xts)
library(ggplot2)
library(lubridate)
library(zoo)
library(xts)
```

```r
##Data Importation
file_name="LSTM-Multivariate_pollution.csv"
data <- read.csv(file_name,header=T)
#Checking size of data column in the dataset
length(data$date)
```

```r
show(data)
```

```r
# Date
data$date <- as.POSIXct(data$date, format = "%Y-%m-%d %H:%M:%S",by="hour")
```

## Hourly

```r
ts_data <- ts(data$pollution, frequency = 24)
```

```r
str(ts_data)
```

```r
autoplot(ts_data, xlab="Time", ylab="Pollution",
     main="Pollution in time")
```

### Trend

```r
autoplot(ts_data, facets = TRUE) +
  geom_smooth() +
  labs("International arrivals to Australia",
       y = "Arrivals (in thousands)",
       x = NULL)
```

### Seasonality

```r
ggseasonplot(ts_data, year.labels=FALSE, continuous=TRUE)
```

```r
ggsubseriesplot(ts_data)
```

```r
ggsubseriesplot(ts_daily)
```

```r
# Load the urca package
library(urca)


# Perform the KPSS test
kpss_test <- ur.kpss(ts_data)

# Print the test results
summary(kpss_test)
```

### autocorrelation

```r
acf(ts_data , lag=24)
```

```r
additive.decomposition<-decompose(ts_data,type=c("additive"))
plot(additive.decomposition)
```

```r
daily_ts <- aggregate(ts_data, FUN = mean)
```

```r
ts_daily <- ts(daily_ts, frequency = 7)
```

```r
plot(ts_daily, xlab="Time daily", ylab="Pollution",
     main="Pollution daily",
     col="purple4",lwd=2)
```

```r
additive.decomposition<-decompose(ts_daily)
plot(additive.decomposition)
```

```r
weekly_ts <- ts(ts_daily, frequency = 365.25/7)

length(weekly_ts)
```

```r
acf(weekly_ts , lag=7)
```

```r
plot(weekly_ts)
```

```r
additive.decomposition<-decompose(weekly_ts)
plot(additive.decomposition)
```

```r
install.packages("tseries")
library(tseries)
adf.test(daily_ts)
```

```r
acf(daily_ts)
Box.test(daily_ts)
```

```r
acf(daily_ts, 7)
pacf(daily_ts, 7)
```

```r
install.packages("forecast")
library(forecast)

ARIMA=auto.arima(daily_ts,d=1,trace=TRUE)
```

```r
# our best model is  ARIMA (0 , 1 , 3)

Model.res<-arima(daily_ts,order = c(0, 1, 3))
adf.test(Model.res$residuals)
summary(Model.res)
checkresiduals(Model.res)
```

```r
d.sp.ts.7<-diff(daily_ts,7,1)
acf(d.sp.ts.7,50)
pacf(d.sp.ts.7,50)
```

```r
arima(daily_ts, order = c(0, 1, 3),seasonal = list(order = c(0,1,2), period = 7))$sigma2
arima(daily_ts, order = c(0, 1, 3),seasonal = list(order = c(1,1,2), period = 7))$sigma2 #the selected model
arima(daily_ts, order = c(0, 1, 3),seasonal = list(order = c(2,1,0), period = 7))$sigma2
arima(daily_ts, order = c(0, 1, 3),seasonal = list(order = c(2,1,1), period = 7))$sigma2
```

```r
Model.res.s<-arima(daily_ts, order = c(0, 1, 3),seasonal = list(order = c(2,1,1), period = 7))
adf.test(Model.res.s$residuals)
summary(Model.res.s)
checkresiduals(Model.res.s)
```

```r
install.packages("aTSA")
library(aTSA)
ts.diag(Model.res.s)
```
